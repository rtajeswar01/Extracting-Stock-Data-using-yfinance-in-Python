# Extracting Stock Data in Python

## Extracting Stock Data of Apple Inc

```python
#Import the libraries 
import yfinance as yf
import pandas as pd
```

```python
# Specify the stock symbol

symbol = "AAPL"  # Apple Inc.
```

```python
# Create a Ticker object with the stock symbol
ticker = yf.Ticker(symbol)

# Get the stock information as a dictionary
stock_info = ticker.info

# Print the stock information
for key, value in stock_info.items():
    print(key + ": " + str(value))
```

```python
#Print the country
stock_info['country']
```

```python
#Print the zip code
stock_info['zip']
```

### Extracting Share Price

```python
# Get the historical stock data
data = ticker.history(period="1d")

# Extract the latest share price
latest_price = data['Close'].iloc[-1]

print(f"The latest share price of {symbol} is: {latest_price}")
```

```python
# Get the historical stock data for the past 5 trading days
data = ticker.history(period="5d")

# Print the entire DataFrame
print(data)
```

```python
import matplotlib.pyplot as plt
# Get the historical stock data
data = ticker.history(period="1y")

# Extract the closing prices
closing_prices = data['Close']

# Plot the closing prices
plt.figure(figsize=(10, 6))
plt.plot(closing_prices)
plt.title("Apple Stock - Closing Prices")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Extracting Stock Data of Advanced Micro Devices (AMD)

```python
# Create a Ticker object for AMD
amd = yf.Ticker("AMD")
```

```python
# Get the country of the stock
country = amd.info['country']

# Print the country
print("Country:", country)
```

```python
# Get the sector of the stock
sector = amd.info['sector']

# Print the sector
print("Stock:", sector, "sector")
```

```python
# Get the historical stock data
data = amd.history(period="max")

# Extract the volume traded on the first day (first row)
first_day_volume = data.iloc[0]['Volume']

# Print the volume traded on the first day
print("Volume traded on the first day:", first_day_volume)
```

```python
adj_close_prices = data['Close']

# Plot the adjusted closing prices
plt.figure(figsize=(10, 6))
plt.plot(adj_close_prices)
plt.title("AMD Stock - Adjusted Closing Prices")
plt.xlabel("Date")
plt.ylabel("Adjusted Closing Price")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
